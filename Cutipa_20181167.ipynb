### Tratamiento de datos

El objetivo final del trabajo será ver los efectos la preparación para el examen y del nivel educativo de los padres en el rendimiento en matemática del los alumnos. Por lo tanto, primero exploramos cuales son las categorías de estas variables:

```python
pd.value_counts(data['test preparation course'])
```

```python
pd.value_counts(data['parental level of education'])
```

Necesitaremos los datos de nivel educativo de los padres de manera ordenada, por lo tanto, recodificamos las variables categóricas en niveles. El nivel 1 representa el más bajo y el 6, el más alto.

```python
data["ple_int"]=data["parental level of education"].replace(["some high school" , "high school", "some college", "bachelor's degree", "associate's degree", "master's degree"], [1, 2, 3, 4, 5, 6])
pd.value_counts(data['ple_int'])
```

Se muestran las frecuencias de los datos codificados. Además, ple_int hace referencia a al "parental level of education" en variables numéricas.


### Elaboración de gráficos

A pesar de ser requerido solo 1  gráfico, consideré pertinente presentar 2. Primero importamos el paquete altair:

```python
import altair as alt
```

#### Gráfico 1

A continuación, se muestra el efecto del curso de preparación para los exámenes en el puntaje en matemáticas:

```python
range_ = ['#f43f3f','steelblue']

alt.Chart(data).mark_area(
    opacity=0.5,
    interpolate='step'
).encode(
    alt.X('math score:Q', bin=alt.Bin(maxbins=100)),
    alt.Y('count()', stack=None),
    alt.Color('test preparation course:N',
              scale=alt.Scale(range=range_))
).properties(
    title='Math score'
).configure_axis(
    grid=False)
```

Como se puede ver, la mayoría de estudiantes no completaron el curso. Además, queines lo completaron, presentan una ligera ventaja en comparación de quienes no lo completaron.

#### Gráfico 2.

Luego, se presenta la relación entre la educación de los padres y el puntaje en matemáticas de los alumnos, separados en dos grupos. Quienes completaron el curso de preparación y quienes no lo completaron.

```python
range_ = ['#f43f3f','steelblue']

alt.Chart(data).mark_point(color='green').encode(
    x = alt.X("math score"),
    y = alt.Y("ple_int"),
    color = alt.Color("test preparation course", 
            legend=alt.Legend(orient='bottom', titleOrient='left'),
            scale=alt.Scale(range=range_)),
    column = alt.Column("test preparation course")
    ).properties(width=300, height=300)
```

Como se puede apreciar en el primer recuadro, el nivel de educación de los padres parece afectar muy poco o nada los puntajes en matemática de los hijos cuando estos completan el curso de preparación. Sin embargo, para quienes no lo completan, se aprecia una relación positiva. Es decir, mientras mayor sea la eduación de los padres, mayor será el puntaje en matemáticas obtenido por los hijos.
